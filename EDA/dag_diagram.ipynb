**Dag Diagram**
- Easy to use visualisation to map out the relationships between the variables
- These potential relationships were identified in the EDA process

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create an instance/object of a directed graph
dag=nx.DiGraph()
```

```python
type(dag)
```

```python
# add nodes/ variables

dag.add_nodes_from(["Churn","Tenure", "Support Cat", "Age Cat","Total Spend","Payment Delay","Contract Type","Total Usage","Usage Frequency","Last Interaction"])
```

```python
# Add directed edges (causal relationships)
dag.add_edges_from([
    ("Churn", "Tenure"),
    ("Churn", "Total Spend"),
    ("Churn", "Support Cat"),
    ("Support Cat","Age Cat"),
    ("Age Cat","Usage Frequency"),
    ("Age Cat","Tenure"),
    ("Support Cat","Contract Type"),
    ("Churn", "Age Cat"),
    ("Churn", "Payment Delay"),
    ("Churn","Contract Type"),
    ("Churn","Total Usage"),
    ("Churn","Last Interaction"),
    ("Churn","Usage Frequency")
])
```

```python
# Draw the graph
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(dag)  # Position nodes using the spring layout
nx.draw(dag, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", arrowsize=20)
plt.title("DAG: Customer Churn", fontsize=14)
plt.show()
```
